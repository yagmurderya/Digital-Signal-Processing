#### 2. Ayrık-zamanlı işaret oluşturmak

```python
n = np.arange(0, 10) # aralık oluşturuldu
x_d = np.cos(n*np.pi/3+np.pi/2) # işaret tanımlandı
print(x_d)
```

#### 3. Sürekli zamanlı işaret oluşturmak

```python
t = np.linspace(0, 2/50, 1000) # çok sık örnekleyersek sürekliymiş gibi olur
x_c = np.cos(100*np.pi*t - np.pi/5)
```

#### 4. Grafik çizdirmek

```python
# 2. başlıkta oluşturulan ayrık-zamanlı işareti çizdirmek için
plt.stem(n, x_d)   # noktaları ayrık çizdirir
plt.xlabel('n')    # x ekseni isimlendirildi
plt.ylabel('x[n]') # y ekseni isimlendirildi
plt.show()

plt.figure() # yeni figür açıldı

# 3. başlıkta oluşturulan sürekli zamanli işareti çizdirmek için
plt.plot(t, x_c)  # noktalar arasını birleştirerek çizdirir
plt.xlabel('t')    # x ekseni isimlendirildi
plt.ylabel('x(t)') # y ekseni isimlendirildi
plt.show()
```

#### 5. Tek figürde birden fazla işaret çizdirmek

```python
t = np.linspace(0, 2/50, 1000)
x1 = np.cos(100*np.pi*t - np.pi/5)
x2 = 2*np.sin(200*np.pi*t)

plt.plot(t, x1, t, x2)
plt.legend(('x1(t)', 'x2(t)')) # işaretlerin renklerine isim atandı
plt.show()
```

#### 6. İşaretleri ötelemek

```python
n = np.array([0, 1, 2, 3, 4])     # indis vektörü
x3 = np.array([1, 0, 2, -1, 0])   # işaret vektörü

plt.xlabel('n')
plt.ylabel('x[n]')
plt.stem(n, x3)
plt.show()
```

```python
# 1. yol 
# n vektörünü değiştirerek iki örnek sağa ötelemek

n2 = n + 2          # her eleman 2 ile toplandı
plt.title('x[n-2]') # başlık
plt.xlabel('n')
plt.ylabel('x[n]')
plt.stem(n2, x3)
plt.show()
```

```python
# 1. yol 
# işaret vektörünü değiştirerek iki örnek sağa ötelemek

n = np.array([0, 1, 2, 3, 4, 5]) 
x = np.array([1, 0, 2, -1, 0, 0]) 

x_shifted = np.append(np.array([0, 0]), x[:-2]) # = [0, 0, 1, 0, 2 ,-1]

fig = plt.figure() # yeni figür

ax1 = fig.add_subplot(211) # 1. satırın 1. sütununa
ax1.stem(n, x)
ax1.title.set_text('x[n]') # başlık

ax2 = fig.add_subplot(212) # 2. satırın 1. sütununa
ax2.stem(n, x_shifted)
ax2.title.set_text('x[n-2]')
```

### SORU-3

```python
# gerekli kütüphanelerin tanımlanması
import numpy as np # dizi oluşturmak için
import matplotlib.pyplot as plt # grafik çizdirmek için

%matplotlib notebook

pi = np.pi
n = np.arange(0, 16)    # ilk 5 işaretin aralığı
n_6 = np.arange(-10, 5) # 6. işaretin aralığı
```

### I. $x_1[n]=exp(\pi/5)$

```python
x_1 = np.zeros(len(n))    # n vektörü ile aynı uzunlukta x_1 vektörü oluşturuldu.
for k in range(len(n)):
    x_1[k] = np.exp(pi/5) # vektör verilen işarete göre dolduruldu

plt.title('$x_1[n]$')     # başlık 
plt.xlabel('n')           # x ekseni isimlendirildi
plt.stem(n, x_1)          # ayrık zamanda çizdirildi
plt.show()                # gösterildi
```

### II. $x_2[n]=exp(j\pi/5)$

```python
x_2 = np.zeros(len(n), dtype=complex) # n uzunluğunda kompleks vektör oluşturuldu
for k in range(len(n)):
    x_2[k] = np.exp(1j*pi/5)          # vektör verilen işarete göre dolduruldu
    
plt.subplot(2, 2, 1)
plt.title('$x_2[n]$')
plt.ylabel('gerçek')
plt.stem(n, x_2.real)       # işaretin gerçek kısmı çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 2)
plt.title('$x_2[n]$')
plt.ylabel('sanal')
plt.stem(n, x_2.imag)       # işaretin sanal kısmı çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 3)
plt.title('$x_2[n]$')
plt.ylabel('genlik')
plt.stem(n, abs(x_2))       # işaretin genliği çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 4)
plt.title('$x_2[n]$')
plt.ylabel('faz')
plt.stem(n, np.angle(x_2))  # işaretin fazı çizdirildi
plt.xlabel('n')

plt.show()
```

### III. $x_3[n]=exp(j\pi n/5)$

```python
x_3 = np.exp(1j*pi*n/5)     # işaret tanımlandı

plt.subplot(2, 2, 1)
plt.title('$x_3[n]$')
plt.ylabel('gerçek')
plt.stem(n, x_3.real)       # işaretin gerçek kısmı çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 2)
plt.title('$x_3[n]$')
plt.ylabel('sanal')
plt.stem(n, x_3.imag)       # işaretin sanal kısmı çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 3)
plt.title('$x_3[n]$')
plt.ylabel('genlik')
plt.stem(n, abs(x_3))       # işaretin genliği çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 4)
plt.title('$x_3[n]$')
plt.ylabel('faz')
plt.stem(n, np.angle(x_3))  # işaretin fazı çizdirildi
plt.xlabel('n')

plt.show()
```

### IV. $x_4[n]=cos(\pi/5)+jsin(\pi/5)$

```python
x_4 = np.zeros(len(n), dtype=complex)    # n uzunluğunda kompleks vektör oluşturuldu
for k in range(len(n)):
    x_4[k] = np.cos(pi/5) + 1j*np.sin(pi/5)    # vektör verilen işarete göre dolduruldu
    
plt.subplot(2, 2, 1)
plt.title('$x_4[n]$')
plt.ylabel('gerçek')
plt.stem(n, x_4.real)        # işaretin gerçek kısmı çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 2)
plt.title('$x_4[n]$')
plt.ylabel('sanal')
plt.stem(n, x_4.imag)        # işaretin sanal kısmı çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 3)
plt.title('$x_4[n]$')
plt.ylabel('genlik')
plt.stem(n, abs(x_4))        # işaretin genliği çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 4)
plt.title('$x_4[n]$')
plt.ylabel('faz')
plt.stem(n, np.angle(x_4))   # işaretin fazı çizdirildi
plt.xlabel('n')

plt.show()
```

### V. $x_5[n]=cos(\pi n/5)+jsin(\pi n/5)$

```python
x_5 = np.cos(pi*n/5) + 1j*np.sin(pi*n/5)  # işaret tanımlandı

plt.subplot(2, 2, 1)
plt.title('$x_5[n]$')
plt.ylabel('gerçek')
plt.stem(n, x_5.real)      # işaretin gerçek kısmı çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 2)
plt.title('$x_5[n]$')
plt.ylabel('sanal')
plt.stem(n, x_5.imag)      # işaretin sanal kısmı çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 3)
plt.title('$x_5[n]$')
plt.ylabel('genlik')
plt.stem(n, abs(x_5))      # işaretin genliği çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 4)
plt.title('$x_5[n]$')
plt.ylabel('faz')
plt.stem(n, np.angle(x_5))  # işaretin fazı çizdirildi
plt.xlabel('n')

plt.show()
```

### VI. $x_6[n]=cos(\pi n/5)+jsin(\pi n/5)$

```python
x_6 = np.cos(pi*n_6/5) + 1j*np.sin(pi*n_6/5)  # işaret tanımlandı

plt.subplot(2, 2, 1)
plt.title('$x_6[n]$')
plt.ylabel('gerçek')
plt.stem(n_6, x_6.real)      # işaretin gerçek kısmı çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 2)
plt.title('$x_6[n]$')
plt.ylabel('sanal')
plt.stem(n_6, x_6.imag)      # işaretin sanal kısmı çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 3)
plt.title('$x_6[n]$')
plt.ylabel('genlik')
plt.stem(n_6, abs(x_6))      # işaretin genliği çizdirildi
plt.xlabel('n')

plt.subplot(2, 2, 4)
plt.title('$x_6[n]$')
plt.ylabel('faz')
plt.stem(n_6, np.angle(x_6))  # işaretin fazı çizdirildi
plt.xlabel('n')

plt.show()
```
